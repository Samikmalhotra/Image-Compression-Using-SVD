```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from matplotlib.image import imread
%matplotlib inline
from sklearn.preprocessing import StandardScaler

for i in range(1,5):
    #Showing Image    
    image = imread('pneumonia_'+str(i)+'.jpeg')
    a = image[:,:]
    print(image.shape)
    plt.imshow(image,cmap='gray_r')
    plt.show()
    
    #Compressing image using SVD
    arr=a      
    arr=StandardScaler().fit_transform(arr)
    R,G,B = np.linalg.svd(arr,full_matrices = False)
    G = np.diag(G)
    print(a.shape)
    print(R.shape)
    print(G.shape)
    print(B.shape)
    j = 1
    
    for k in (10,50,75,100):
        approx = R[:,:k] @ G[0:k,:k] @ B[:k,:]
        plt.figure(j)
        j = j+1
        print('k=' + str(k)+'_pneumonia_'+str(i)+'.jpg')
        img = plt.imshow(approx)
        img.set_cmap('gray')
        plt.axis('off')
        plt.title('k = ' + str(k))
        plt.savefig('k=' + str(k)+'_pneumonia_'+str(i)+'.jpg')
        print(approx.shape)
        plt.show()
        

    

    
    
```
